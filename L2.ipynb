###Задание1

```python
import numpy as np
import pandas as pd
authors = pd.DataFrame({'author_id':[1, 2, 3],
                        'author_name':['Тургенев', 'Чехов', 'Островский']}, columns=(['author_id', 'author_name'])) 
```

```python
authors
```

```python
book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                     'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                     'price':[450, 300, 350, 500, 450, 370, 290]}, columns=(['author_id', 'book_title', 'price']))
```

```python
book
```

###Задание2

```python
authors_price = pd.merge(authors, book, on='author_id', how='outer')
```

```python
authors_price
```

###Задание3

```python
top5 = authors_price.nlargest(5, 'price')
```

```python
top5
```

###Задание4

```python
authors_stat = authors_price.copy()
authors_stat = authors_stat.drop('author_id', axis=1)
authors_stat = authors_stat.drop('book_title', axis=1)
authors_stat = authors_stat.groupby('author_name').agg({'price':['max', 'min', 'mean']}).round(2)
authors_stat = authors_stat.rename(columns={'max': 'max_price', 'min': 'min_price', 'mean': 'mean_price'})

#df.assign(mean_a=df.a.mean(), mean_b=df.b.mean() изучить дополнительно
```
